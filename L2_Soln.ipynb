```python
import warnings
warnings.simplefilter(action='ignore', category = FutureWarning)

import pandas as pd
import numpy as np
```

##### **Part 'a'**

```python
# read the data from csv file (with first column as the index) & store it in a dataframe
df1 = pd.read_csv("data/data1.csv", index_col = 0)

df1 # displays the dataframe 1
```

```python
# read the data from csv file (with second column as the index) & store it in a new dataframe
df2 = pd.read_csv("data/data2.csv", index_col = 1)

df2 # dataframe 2
```

##### **Part 'b'**

```python
# merging two dataframes (on same columns)
df3 = pd.concat([df1, df2])

df3 # dataframe 3
```

##### **Part 'c'**

```python
# importing new dataframe from csv file (with first column as the index)
df4 = pd.read_csv("data/data3.csv", index_col = 0)

df4 # dataframe 4
```

```python
# merging two dataframes (on different columns)
df5 = pd.concat([df3, df4], axis = 1)

# rearrange rows in ascending order of the index [extra step]
df5 = df5.sort_index()

df5 # dataframe 5
```

##### **Part 'd'**

```python
# read the data from json file & store it in a new dataframe
df6 = pd.read_json("data/data.json")

df6 # dataframe 6
```

```python
# merging two dataframes on same columns & ignoring their previous indexes (to generate new indexes for merged dataframe)
df7 = pd.concat([df5, df6], ignore_index = 1)

df7 # dataframe 7
```

##### **Part 'e'**

```python
# changing the datatype of all columns to 'float' & assign NaN to non-numeric values 
df7 = df7.apply(pd.to_numeric, errors = "coerce")

df7 # updated dataframe 7
```

##### **Part 'f'**

```python
# filling NaN values with mean of their respective columns
df7.fillna(df7.mean(), inplace = True)

# rounding off the values to 2 decimal places (for entire dataframe)
df7 = df7.round(2)

df7 # final dataframe 7
```

```python
# saving the final dataframe to a new csv file
df7.to_csv("data/newdata.csv")
```

**Side Notes:**
<br>&emsp;
**1)** Use 'inplace = True' argument only on entire dataframe, not on individual columns. Useful when working with very large datasets.
<br>&emsp;
**2)** Instead of using sort_index(), we can generate new indexes using reset_index() method if row order isn't important
